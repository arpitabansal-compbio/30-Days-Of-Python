```python
# Section : Advanced String Architecture & Method Chaining

# 1. Capture messy, unformatted investigator input
raw_input = "   arpita bansal   "

# 2. Method Chaining: Strip spaces AND apply Title Case in one single line!
cleaned_name = raw_input.strip().title()

# 3. Split Protocol: Chop the clean name into separate variable boxes
first_name, last_name = cleaned_name.split()

# 4. Print clean structured diagnostic variables using F-strings
print(f"First Name Box: {first_name}")
print(f"Last Name Box:  {last_name}")
```
