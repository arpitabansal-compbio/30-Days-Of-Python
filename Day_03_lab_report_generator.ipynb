```python
"""
GENERATING LAB REPORT
"""
# 1.ask for inputs and store them in variable boxes
scientist_name = input("Enter your name: ")
total_samples = input("Enter total DNA samples analyzed: ")
passed_samples = input("Enter number of samples that passed quality control: ")
# 2.create a sepeartor line using print with 'sep' parameter
print("=", "=", "=", "=", sep="-")
# 3.Print the title and details
print("DAILY LAB REPORT ")
print("Scientist: " + scientist_name)
print("Total samples: " + total_samples)
print("Passed samples: " + passed_samples)
# 4. create another seperator line
print("=", "=", "=", "=", sep="-")
# 5. print a closing message (using 'end' paramter for formatting)
print("Report generated successfully.", end="")

```
